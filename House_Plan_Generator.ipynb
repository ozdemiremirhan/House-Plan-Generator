```python
# Stable Diffusion ve gerekli kütüphaneleri yükleyin
!pip install diffusers
!pip install transformers
!pip install accelerate
!pip install ftfy

import torch
from diffusers import StableDiffusionPipeline

# Modeli yükleyin
model_id = "CompVis/stable-diffusion-v1-4"
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = StableDiffusionPipeline.from_pretrained(model_id).to(device)
```

```python
# Kullanıcıdan oda bilgilerini al
yatak_odasi_sayisi = int(input("Yatak odası sayısı: "))
salon_sayisi = int(input("Salon sayısı: "))
mutfak_sayisi = int(input("Mutfak sayısı: "))
banyo_sayisi = int(input("Banyo sayısı: "))
tuvalet_sayisi = int(input("Tuvalet sayısı: "))
oturma_odasi_sayisi = int(input("Oturma odası sayısı: "))
calisma_odasi_sayisi = int(input("Çalışma odası sayısı: "))
kiler_sayisi = int(input("Kiler sayısı: "))
camasir_odasi_sayisi = int(input("Çamaşır odası sayısı: "))

# Prompt oluştur
prompt = (f"A floor plan with {yatak_odasi_sayisi} bedrooms, "
          f"{salon_sayisi} living rooms, {mutfak_sayisi} kitchens, "
          f"{banyo_sayisi} bathrooms, {tuvalet_sayisi} toilets, "
          f"{oturma_odasi_sayisi} sitting rooms, {calisma_odasi_sayisi} study rooms, "
          f"{kiler_sayisi} storage rooms, and {camasir_odasi_sayisi} laundry rooms. Architectural accurate, super realistic.")
```

```python
# Görüntü oluştur
import cv2
def generate_image(prompt):
    if device == "cuda":
        with torch.autocast("cuda"):
            result = pipe(prompt, guidance_scale=7.5)
    else:
        result = pipe(prompt, guidance_scale=7.5)

    image = result.images[0]
    return image

# Görüntüyü oluşturduktan sonra
image = generate_image(prompt)

# Görüntüyü dosyaya kaydetme
image.save("generated_image.jpg")  # Dosyayı JPEG formatında kaydetmek için



print("Görüntü başarıyla kaydedildi.")
```
